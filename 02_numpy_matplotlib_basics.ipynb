```python
import matplotlib.pyplot as plt
import numpy as np

# 100 adet rastgele veri üret (Model çıktısı gibi düşün)
confidence_scores = np.random.uniform(low=0.1, high=1.0, size=100)

# Bu skorların grafiğini çizdir
plt.figure(figsize=(8, 4))
plt.hist(confidence_scores, bins=10, color="skyblue", edgecolor="black")
plt.title("Model Güven Skorları Dağılımı")
plt.xlabel("Güven Skoru")
plt.ylabel("Frekans")
plt.grid(True)
plt.show()
```
